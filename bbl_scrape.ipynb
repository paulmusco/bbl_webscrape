```python
soup = BeautifulSoup(driver.page_source)
```

```python
soup
```

```python
# Get table column names
col_headers = soup.find_all(class_ = 'bbl-table-cell relative sticky bg-white z-40 top-0 border-2 border-transparent font-black text-xl uppercase italic min-w-max my-0 px-4 py-2 border-b-gray-silver active')

column_names = []
for col in col_headers:
    column_names.append(col.text)
```

```python
print(column_names)
```

```python
# Get table content
all_content = soup.find_all(class_ = 'bbl-table-column paragraph relative my-0 px-4 py-2 min-w-max w-full max-w-none whitespace-nowrap z-0 text-center')

table_content_list = []
for i in all_content:
    table_content_list.append(i.text)
```

```python
# Split long list into separate lists for each row
separated_content_list = []

for i in range(0, len(table_content_list), len(column_names)+1):
    separated_content_list.append(table_content_list[i:i + (len(column_names)+1)])
```

```python
len(separated_content_list[0])
```

```python
# Remove first element of each sub-list
for lst in separated_content_list:
    del lst[0]
```

```python
len(separated_content_list[0])
```

```python
# Create DF
df = pd.DataFrame(separated_content_list, columns = column_names)#, dtype = float)
df
```

```python
# Export DF to CSV
df = df.to_csv('bbl_team_statistics_2020_2021.csv', index=False)
```
